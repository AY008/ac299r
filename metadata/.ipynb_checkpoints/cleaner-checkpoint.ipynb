```python
profiles = pd.read_csv('brides_marathi_full.csv')
```

```python
profiles['age2_0']=profiles['age2'].map(lambda x: str(x)[0:2] if x else "")
profiles['age2_0']=profiles['age2_0'].map(lambda x: int(x) if x != 'na' else None)

profiles['age2_1']=profiles['age2'].map(lambda x: str(x)[6:8] if x else "")
profiles['age2_1']=profiles['age2_1'].map(lambda x: int(x) if x != '' else None)
```
