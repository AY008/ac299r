```python
import string

def getweight(wt): 
    weight = str(wt)
    
    s = [int(s) for s in weight.split() if s.isdigit()]
    
    if len(s) == 2:
        if 'Pound' not in weight:
            toReturn = int((s[0] + s[1]) / 2)
        else:
            toReturn = int((0.5 * (s[0] + s[1]))/2.2)
    elif len(s) == 1:
        if len(s) == 1 and 'Pound' not in weight:
            toReturn = int(s[0])
        elif len(s) == 1 and 'Pound' in weight:
            toReturn = int(s[0] / 2.2)
    else:
        return np.nan
            
    if toReturn != 0 and toReturn >= 40 and toReturn <= 140:
        return toReturn
    else:
        return np.nan  
```

```python
def from_one(height):
    feet, inches = 0, 0
    if "to" not in height:
        if 'ft' in height:
            feet = int(height[0])
        
        if 'in' in height:
            inches = int(height[4:-2])
            
        return feet * 12 + inches



def convert(height):
    height = str(height)
    num = 0
    
    if height == "Doesn't Matter" or height == "Doesn't Matter to ":
        num = 0
    elif height == "5ft to Doesn't Matter":
        num = 60
    elif "to Doesn't Matter" in height:
        num = from_one(height[:-18])
    elif "Doesn't Matter to " in height:
        num = from_one(height[18:])
    
    if "to" not in height:
        num = from_one(height)
        
    if "to" in height:
        heights = height.split(" to ")
        num = int(from_one(heights[0]) + from_one(heights[1]) / 2)
        
    if (num >= 48 and num <= 84):
        return num
    else:
        return np.nan
```

```python
def getloc(details):
    details = str(details)
    detailslen = len(details.split('<br />'))
    details2 = details.split('<br />')[detailslen-1]
    
    if details2.count(",") == 2:
        wholestring = details2.split(',', 1)[1].strip()
    else:
        wholestring = details2.strip()
        
    finalsize = len(wholestring.split(',', 1))

    if finalsize == 2:
        return wholestring.split(',', 1)[0].strip(), wholestring.split(',', 1)[1].strip()
    elif finalsize == 1:
        return "", wholestring.split(',', 1)[0].strip()
    else:
        return "", ""
```
